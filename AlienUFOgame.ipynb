```python


#list of words, random, correct and incorrect guesses, etc
f = open("nouns.txt", "r")
words = f.readlines()          
words = [word.rstrip('\n') for word in words]
    
while True: 

        
    random_word = random.choice(words)
    count = 1
    guess_list = []
    incorrect_list = []
    guess_word = " "
    correct_letters = 0
    won = False

    while True:
        # first ask the player to guess a letter
        original = str(input("Please guess a letter: "))
        guess = original.upper()
        if guess == ".": 
            break
        # this checks if the letter is in the word"
        if guess not in guess_list:
            
            
            if correct_letters == len(random_word) - 1:
                print("Correct! You're closer to cracking the code")
                print(f'The codeword is: {random_word}')
                won = True
                break
                
            if guess in random_word.upper():
                  print("Correct! You're closer to cracking the code")
                
                
                    
            else:
                count += 1
                print("Incorrect! The tractor beam pulls the person in further")       
                incorrect_list.append(guess)

                if count > 6: # You only have 6 lives until the person gets sucked up
                    print("You have reached maximum attempts and now there is an alien abduction")
                    break
            
            
            
            print(graphics[count-1]) # graphics of person going into spaceship
            guess_list.append(guess)
            print(f"Incorrect Guesses: {' '.join(incorrect_list)}")
            
            guess_word = ""
            correct_letters = 0
            #loop through all letters to guess, if the letters in the codeword, replace dash or try again
            for i in range(len(random_word)):

                if random_word[i].upper() in "".join(guess_list).upper():
                    guess_word += random_word[i]
                    correct_letters += 1
                
                else:
                    guess_word += "_"

                guess_word += " "


            print("\nCodeword: ")
            print(guess_word)
        else:
            print("Incorrect guessed {}. Try again".format(guess))
            print(set(guess_list))

    # Ask the player if they want to play again, if not game ends. If yes, start with a new word (include delighter!!!)
    if won:
        choice = str(input("You won and saved the human. Would you like to play again (Y/N)? "))
    else:
        choice = str(input("If at first you don't succeed, try, try again!. Would you like to play again (Y/N)? "))   
    choice = choice.upper()
    if choice == "N": 
        break
        while True: 
            if choice == "Y":
                guess = str(input("Please guess a letter: "))
            else:
                print("It was great playing with you!")


```
